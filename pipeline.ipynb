```python
# 口语化改写pipeline
# 用于把所有生成的query问题进行拟人化
import random

humanlike_prompt="我这边有一个用户的具体人设信息，与其向AI提问的意图问题。请根据该意图和用户的人设，生成以这个用户的说话口吻会使用的询问方式。"
strict='你必须严格遵循原始意图的所有内容，仅改为接近口语的格式。不要增加、减少或修改意图。'
non_strict='你可以根据用户人设应有的知识水平和说话习惯适当丰富或减少提问内容中的细节信息，增加或减少部分意图。'

def human_like(q,charactor=None,strict_bool=False):
    if charactor==None:
        charactor=random.choice(charactors)['core_persona']
    if strict_bool:
        return client.chat(humanlike_prompt+strict+'\n\n原始意图问题：\n'+q+'\n\n用户人设：\n'+charactor)
    else:
        return client.chat(humanlike_prompt+non_strict+'\n\n原始意图问题：\n'+q+'\n\n用户人设：\n'+charactor)
```

```python
# 获取网站主题类型
import requests

sys_prompt='你是一个网页前端设计专家，有着网页前端设计的丰富经验，能够从网页的代码角度判断其核心主题。用户正尝试让你分析不同网站的主题要素、细分领域、用户需求等，以及其是否有明显定制化的设计。你将被给到一个网页的完整html代码，而你需要以json格式{"主题":"xxx","细分领域":"xxx","用户需求":"xxx","定制化设计":"xxx"}返回结果。'

def check_one_site(url):
    res=requests.get(url)
    result=client.chat([{'role':'system','content':sys_prompt},{'role':'user','content':res.text}])
    return result

site_types=[]
sites=['https://www.baidu.com','https://www.taobao.com','https://www.hao123.com/']
for site in sites:
    site_types.append(check_one_site(site))

print(site_types)
```
